```python
#the filter procduces a pandas series of True and False values # We just want to choose Toyota Sienna cars

car_filter = df['car_type'] == 'Toyota Sienna'
```

```python
car_filter.head()
```

```python
df[car_filter].head() # Cheacking the top values
```

```python
df.loc[car_filter, :] # checking all values if they all have Toyota Sienna cars
```

```python
df['car_type'].value_counts() #The values did not change because we did not add filter data frame to the original data frame    
```

```python
df = df.loc[car_filter, :] # filter dataframe to get a DataFrame of only 'Toyota Sienna
```

```python
df['car_type'].value_counts() # Now instead of having all values, we just have values with Toyota Sienna
```
